```python
def getRanks4(gene_groups, h, m0, m1, order):
    expr = []
    row_labels = []
    row_ids = []
    row_numhi = []
    ranks = []
    g_ind = 0
    counts = []
    for s in gene_groups:
        count = 0
        avgrank = [0 for i in order]
        for gn in s:
          for id in h.getIDs(gn):
            e = h.getExprData(id);
            if len(e) != (h.getEnd()+1):
                continue
            v = np.array([float(e[i]) if e[i] != "" else 0 for i in order])
            if (np.max(v) - np.min(v)) <= 0:
                continue
            v1 = np.mean([float(e[i]) if e[i] != "" else 0 for i in m1])
            v0 = np.mean([float(e[i]) if e[i] != "" else 0 for i in m0])
            t = [v1, v1, v1, (v0 + v1)/2]
            #print(id)
            #print(v)
            #print(t[3])
            #print(np.std(v))
            te = []
            for i in range(len(order)):
                if e[order[i]] == "":
                    v1 = - t[3] / 3;
                else:
                    v1 = (float(e[order[i]]) - t[3]) / 3;
                if np.std(v) > 0:
                    v1 = v1 / np.std(v)
                avgrank[i] += v1
                te.append(v1)
            #print(avgrank)
            expr.append(te)
            row_labels.append(h.getSimpleName(id))
            row_ids.append(id)
            v1 = [g_ind, sum(v > t[3])]
            if g_ind > 3:
                v1 = [g_ind, sum(v <= t[3])]
            else:
                v1 = [g_ind, sum(v > t[3])]
            row_numhi.append(v1)
            count += 1
            #if count > 200:
            #    break
        ranks.append(avgrank)
        g_ind += 1
        counts += [count]
    print(counts)
    return ranks, row_labels, row_ids, row_numhi, expr

def H1(ana):
    dir1 = "/booleanfs2/sahoo/Data/Macrophage/BN/"
    lh1 = hu.getHash(dir1 + "paper/dna-repair.txt", 0)
    keys = list(lh1.keys())
    keys = ['BER2', 'DSBR', 'MMR', 'BER/SSBR', 'OTHERS', 'BER', 'NER']
    keys = ['DSBR', 'MMR', 'NER']
    keys = ['BER']
    l1 = [hu.uniq(lh1[k][1:]) for k in keys]
    print(keys)
    #print l1
    l1 = [ ['Neil2', 'Neil1', 'Neil3', 'Ogg1', 'Nthl1']]
    l1 = [ ['NEIL2', 'NEIL1', 'NEIL3', 'OGG1', 'NTHL1']]
    genes = [i for k in l1 for i in k]
    m0 = [i for i in ana.h.aRange() if ana.aval[i] == 0]
    m1 = [i for i in ana.h.aRange() if ana.aval[i] == 1]
    ana.order = []
    for k in range(9):
        mk = [i for i in ana.h.aRange() if ana.aval[i] == k]
        ana.order += mk
    col_labels = [ana.h.headers[i] for i in ana.order]
    ranks, row_labels, row_ids, row_numhi, expr = getRanks4(l1, ana.h, m0, m1, ana.order)
    f_ranks = ranks[0]
    i1 = []
    index = []
    for k in range(9):
        mk = [i for i in ana.h.aRange() if ana.aval[i] == k]
        i1 += [ana.order[i] for i in np.argsort(f_ranks) if ana.order[i] in mk]
        index += [i for i in np.argsort(f_ranks) if ana.order[i] in mk]
    cval = np.array([[ana.aval[i] for i in i1]])
    ind_r = np.array(sorted(range(len(row_labels)),
        key=lambda x: (row_numhi[x][0], row_numhi[x][1])))
    row_labels = [row_labels[i] for i in ind_r]
    row_ids = [row_ids[i] for i in ind_r]
    data = np.array([expr[i] for i in ind_r])[:,index]
    ofile = "figures/heatmap-test.pdf"
    params = {'dx': 70, 'dy': 50, 'spaceAnn': len(ana.order)/len(ana.atypes),
              'tAnn': 0.1, 'widthAnn':0.5,
              'sy': 0.9, 'thr': 1, 'w': 8, 'h': 8, 'rowlabels':1,
              'genes': genes, 'atypes': ana.atypes, 'cval': cval,
              'tl': 6, 'tw': 0.25, 'ts': 10, 'tsi': -10}
    ax,bp = bone.plotHeatmap(ofile, data, col_labels, row_labels, params)
    return ax,bp
ana = bone.IBDAnalysis()
ana.getZhang2020(2)
ax, bp = H1(ana)
```

```python
def plotKMF(df, gene, thr):
    fig,ax = bone.plt.subplots(figsize=(7,5), dpi=100)
    time1 = [45 - int(df['time'][i]) for i in df.index if float(df['expr'][i])> thr]
    status1 = [int(df['status'][i]) if int(df['time'][i]) != 0 else 0 for i in df.index if float(df['expr'][i]) > thr]
    from lifelines import KaplanMeierFitter
    kmf1 = KaplanMeierFitter()
    kmf1.fit(time1,event_observed=status1, label=gene + ' high (n=' + str(len(time1)) + ')')
    ax = kmf1.plot_cumulative_density(ax=ax, ci_show=False)
    time2 = [45 - int(df['time'][i]) for i in df.index if float(df['expr'][i])<= thr]
    status2 = [int(df['status'][i]) if int(df['time'][i]) != 0 else 0
               for i in df.index if float(df['expr'][i]) <= thr]
    kmf2 = KaplanMeierFitter()
    kmf2.fit(time2,event_observed=status2, label=gene + ' low (n=' + str(len(time2)) + ')')
    ax = kmf2.plot_cumulative_density(ax=ax, ci_show=False)
    from lifelines.plotting import add_at_risk_counts
    kmfs = [kmf1, kmf2]
    add_at_risk_counts(*kmfs, ax=ax)
    ax.set_xlabel("Days in hospital")
    ax.set_ylabel("Cumulative Probability of Discharge")
    from lifelines.statistics import logrank_test
    from matplotlib.legend import Legend
    #print(hu.uniq(time1), hu.uniq(time2))
    res = logrank_test(time1, time2, status1, status2, alpha=.99)
    leg = Legend(ax, [], [], title = "p = %.2g" % res.p_value,
                       loc='upper right', frameon=False)
    ax.add_artist(leg);
    return

def plotAgeGender(df):
    df['A+G'] = ''
    labels = ["M $>=$ 40", "M $<$ 40", "F $>=$ 40", "F $<$ 40"]
    c1 = df['state'] == 'COVID-19'
    c2 = np.logical_and(df['age'] >= 40, df['gender'] == 'male')
    df.loc[c2, 'A+G'] = labels[0]
    c3 = np.logical_and(df['age'] < 40, df['gender'] == 'male')
    df.loc[c3, 'A+G'] = labels[1]
    c4 = np.logical_and(df['age'] >= 40, df['gender'] == 'female')
    df.loc[c4, 'A+G'] = labels[2]
    c5 = np.logical_and(df['age'] < 40, df['gender'] == 'female')
    df.loc[c5, 'A+G'] = labels[3]
    c2 = np.logical_and(c1, c2)
    c3 = np.logical_and(c1, c3)
    c4 = np.logical_and(c1, c4)
    c5 = np.logical_and(c1, c5)

    kmfs = []
    time_ar = []
    status_ar = []
    ax = None
    fig,ax = bone.plt.subplots(figsize=(7,5), dpi=100)
    v = [c2, c3, c4, c5]
    for k in range(4):
        c = v[k]
        time1 = [45 - int(df[c]['time'][i]) for i in df[c].index]
        status1 = [int(df[c]['status'][i]) if int(df[c]['time'][i]) != 0 else 0 for i in df[c].index]
        print(time1, status1)
        from lifelines import KaplanMeierFitter
        kmf1 = KaplanMeierFitter()
        kmf1.fit(time1,event_observed=status1, label=labels[k] + ' (n=' + str(len(time1)) + ')')
        if ax is None:
            ax = kmf1.plot_cumulative_density(ci_show=False)
        else:
            ax = kmf1.plot_cumulative_density(ax=ax, ci_show=False)
        kmfs += [kmf1]
        time_ar += [time1]
        status_ar += [status1]

    from lifelines.plotting import add_at_risk_counts
    add_at_risk_counts(*kmfs, ax=ax)
    ax.set_xlabel("Days in hospital")
    ax.set_ylabel("Cumulative Probability of Discharge")
    from lifelines.statistics import logrank_test
    from matplotlib.legend import Legend
    #print(hu.uniq(time1), hu.uniq(time2))
    res = logrank_test(time_ar[0] + time_ar[1] + time_ar[2], time_ar[3],
                       status_ar[0] + status_ar[1] + status_ar[2], status_ar[3], alpha=.99)
    leg = Legend(ax, [], [], title = "p = %.2g" % res.p_value,
                       loc='lower right', frameon=False)
    ax.add_artist(leg);
    return
```
