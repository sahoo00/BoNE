```python
import bone
reload(bone)
import Datasets
reload(Datasets)

def plotTitleBar(cval, atypes, params):
    dpi = 100
    if 'dpi' in params:
        dpi = params['dpi']
    w,h = (5, 0.8)
    if 'w' in params:
        w = params['w']
    if 'h' in params:
        h = params['h']
    color_sch1 = ["#3B449C", "#B2509E","#EA4824"]
    color_sch1 = ["#00CC00", "#EFF51A","#EC008C", "#F7941D", "#808285",
            'cyan', 'blue', 'black', 'green', 'red']
    if 'acolor' in params:
        color_sch1 = params['acolor']
    if 'cval' in params:
        cval = params['cval']

    ax = None
    if 'ax' in params:
        ax = params['ax']
    if ax is None:
        fig = bone.plt.figure(figsize=(w,h), dpi=dpi)
        ax = fig.add_subplot(1, 1, 1)
    nAt = len(cval[0])
    extent = [0, nAt, 0, 5]
    ax.axis(extent)
    cmap = bone.colors.ListedColormap(color_sch1)
    boundaries = range(len(color_sch1) + 1)
    norm = bone.colors.BoundaryNorm(boundaries, cmap.N, clip=True)
    #ax.imshow(cval, interpolation='none', cmap=cmap, \
    #                  norm=norm, extent=extent, aspect="auto")
    y = [0, 5]
    x = bone.np.arange(nAt + 1)
    ax.pcolormesh(x, y, cval, cmap=cmap, norm=norm, zorder=-1.0)
    ax.set_xticklabels([])
    ax.set_yticklabels([])
    ax.tick_params(top=False, left=False, bottom=False, right=False)
    ax.set_xticks(bone.np.arange(0, nAt, 1))
    ax.grid(which='major', alpha=0.2, linestyle='-', linewidth=0.5,
            color='black', zorder=2.0)
    for edge, spine in ax.spines.items():
                spine.set_visible(False)
    divider = bone.make_axes_locatable(ax)
    width = bone.axes_size.AxesX(ax, aspect=1./20)
    spaceAnn = 70
    widthAnn = 3
    tAnn = 1
    if 'spaceAnn' in params:
        spaceAnn = params['spaceAnn']
    if 'widthAnn' in params:
        widthAnn = params['widthAnn']
    if 'tAnn' in params:
        tAnn = params['tAnn']
    pad = bone.axes_size.Fraction(0.1, width)
    lax = divider.append_axes("top", size="100%", pad="20%", frame_on=False)
    lax.axison = False
    lax.axis(extent)
    lax.set_xticklabels([])
    lax.set_yticklabels([])
    lax.grid(False)
    lax.tick_params(top=False, left=False, bottom=False, right=False)
    if 'atypes' in params:
        atypes = params['atypes']
    bone.barTop(lax, atypes, color_sch1, params)
    return ax

bone.plotTitleBar = plotTitleBar

pdf = bone.getPDF("figure-2.pdf")

ana = bone.IBDAnalysis()

wt1, l1 = mut.getCls13()
ana.getSharma2022Mac(4)
ana.orderData(l1, wt1)
fig = mut.plotViolinBar(ana, "C#13 Labor")
fig.axes[1].set_xlim([-30, 40])
pdf.savefig(transparent=True, bbox_inches = 'tight')

wt1, l1 = mut.getCls13()
ana.getSharma2022Mac(3)
ana.orderData(l1, wt1)
fig = mut.plotViolinBar(ana, "C#13 Labor=No Placenta")
ana.printAllPvals()
fig.axes[1].set_xlim([-30, 40])
pdf.savefig(transparent=True, bbox_inches = 'tight')

wt1, l1 = mut.getCls13()
ana.getSharma2022Mac(2)
ana.orderData(l1, wt1)
fig = mut.plotViolinBar(ana, "C#13 Labor=Yes Placenta")
fig.axes[1].set_xlim([-30, 40])
pdf.savefig(transparent=True, bbox_inches = 'tight')

wt1, l1 = mut.getCls13()
ana.getSharma2022Mac(5)
ana.orderData(l1, wt1)
#mut.MacAnalysis.normGene(ana, 'FCGR3A')
fig = mut.plotViolinBar(ana, "C#13 preterm labor=No iMNC BPD")
fig.axes[1].set_xlim([-15, 30])
pdf.savefig(transparent=True, bbox_inches = 'tight')

wt1, l1 = mut.getCls13()
ana.getSharma2022Mac(5)
ana.orderData(l1, wt1)
mut.MacAnalysis.normGene(ana, 'FCGR3A')
fig = mut.plotViolinBar(ana, "C#13 preterm labor=No iMNC BPD CD16")
fig.axes[1].set_xlim([-10, 15])
pdf.savefig(transparent=True, bbox_inches = 'tight')

wt1, l1 = mut.getCls13()
ana.getSharma2022Mac(6)
ana.orderData(l1, wt1)
#mut.MacAnalysis.normGene(ana, 'FCGR3A')
fig = mut.plotViolinBar(ana, "C#13 preterm labor=No cMNC BPD")
fig.axes[1].set_xlim([-15, 30])
pdf.savefig(transparent=True, bbox_inches = 'tight')

wt1, l1 = mut.getCls13()
ana.getSharma2022Mac(6)
ana.orderData(l1, wt1)
mut.MacAnalysis.normGene(ana, 'FCGR3A')
fig = mut.plotViolinBar(ana, "C#13 preterm labor=No cMNC BPD CD16")
fig.axes[1].set_xlim([-10, 15])
pdf.savefig(transparent=True, bbox_inches = 'tight')

wt1, l1 = mut.getCls13()
ana = bone.IBDAnalysis()
ana.getBell2023(5, 'CordBlood')
ana.orderData(l1, wt1)
#mut.MacAnalysis.normGene(ana, 'FCGR3A')
fig = mut.plotViolinBar(ana, "CB")
ana.printAllPvals()
pdf.savefig(transparent=True, bbox_inches = 'tight')

wt1, l1 = mut.getCls13()
ana = bone.IBDAnalysis()
ana.getBell2023(5, 'CordBlood')
ana.orderData(l1, wt1)
mut.MacAnalysis.normGene(ana, 'FCGR3A')
fig = mut.plotViolinBar(ana, "CB CD16")
ana.printAllPvals()
pdf.savefig(transparent=True, bbox_inches = 'tight')

ana.getRyan2019(2)
ana.orderData(l1, wt1)
#mut.MacAnalysis.normGene(ana, 'FCGR3A')
fig = mut.plotViolinBar(ana, "Ryan PBMC")
ana.printAllPvals()
pdf.savefig(transparent=True, bbox_inches = 'tight')

ana.getRyan2019(2)
ana.orderData(l1, wt1)
mut.MacAnalysis.normGene(ana, 'FCGR3A')
fig = mut.plotViolinBar(ana, "Ryan PBMC CD16")
ana.printAllPvals()
pdf.savefig(transparent=True, bbox_inches = 'tight')

ana.getCheng2018()
ana.orderData(l1, wt1)
#mut.MacAnalysis.normGene(ana, 'FCGR3A')
fig = mut.plotViolinBar(ana, "Cheng PBMC")
ana.printAllPvals()
pdf.savefig(transparent=True, bbox_inches = 'tight')

ana.getCheng2018()
ana.orderData(l1, wt1)
mut.MacAnalysis.normGene(ana, 'FCGR3A')
fig = mut.plotViolinBar(ana, "Cheng PBMC CD16")
ana.printAllPvals()
pdf.savefig(transparent=True, bbox_inches = 'tight')

wt1, l1 = mut.getCls13()
ana = bone.IBDAnalysis()
ana.getSun2025BPDII(2, 'pMON')
ana.orderData(l1, wt1)
fig = mut.plotViolinBar(ana, "LungMap pMON")
ana.printAllPvals()
pdf.savefig(transparent=True, bbox_inches = 'tight')

ana.getSun2025BPDII(2, 'iMON')
ana.orderData(l1, wt1)
fig = mut.plotViolinBar(ana, "LungMap iMON")
ana.printAllPvals()
pdf.savefig(transparent=True, bbox_inches = 'tight')

ana.getSun2025BPDII(3)
ana.orderData(l1, wt1)
fig = mut.plotViolinBar(ana, "LungMap Macrophages")
ana.printAllPvals()
pdf.savefig(transparent=True, bbox_inches = 'tight')

ana.getSun2025BPDII(2, 'AM')
ana.orderData(l1, wt1)
fig = mut.plotViolinBar(ana, "LungMap AM")
ana.printAllPvals()
pdf.savefig(transparent=True, bbox_inches = 'tight')

ana.getSun2025BPDII(2, 'IM')
ana.orderData(l1, wt1)
fig = mut.plotViolinBar(ana, "LungMap IM")
ana.printAllPvals()
pdf.savefig(transparent=True, bbox_inches = 'tight')

ana.getSun2025BPDII(4)
ana.orderData(l1, wt1)
fig = mut.plotViolinBar(ana, "LungMap DC")
ana.printAllPvals()
pdf.savefig(transparent=True, bbox_inches = 'tight')

ana.getSun2025BPDII(2, 'pDC')
ana.orderData(l1, wt1)
fig = mut.plotViolinBar(ana, "LungMap pDC")
ana.printAllPvals()
pdf.savefig(transparent=True, bbox_inches = 'tight')

ana.getSun2025BPDII(2, 'cDC1')
ana.orderData(l1, wt1)
fig = mut.plotViolinBar(ana, "LungMap cDC1")
ana.printAllPvals()
pdf.savefig(transparent=True, bbox_inches = 'tight')

ana.getSun2025BPDII(2, 'cDC2')
ana.orderData(l1, wt1)
fig = mut.plotViolinBar(ana, "LungMap cDC2")
ana.printAllPvals()
pdf.savefig(transparent=True, bbox_inches = 'tight')

ana.getSun2025BPDII(2, 'maDC')
ana.orderData(l1, wt1)
fig = mut.plotViolinBar(ana, "LungMap maDC")
ana.printAllPvals()
pdf.savefig(transparent=True, bbox_inches = 'tight')

def adjustFigure(ana, fig):
    rocauc = ana.getROCAUC()
    ax = fig.axes[1]
    #ax.set_xlim([-5, 10])
    ax.set_title(f"ROC-AUC = {rocauc}", fontsize=14)
    ax.set_xlabel("C13 score", fontsize=16)
    ax.tick_params(axis='x', which='major', labelsize=12)
    ax.tick_params(axis='y', which='major', labelsize=16)
    #fig.patch.set_facecolor('#FFF7CD')
    #ax.set_facecolor('#FFF7CD')
    for tobj in ax.texts:
        tobj.set_fontsize(14)
        #tobj.set_color('black')
    #for tobj in fig.findobj(plt.Text):
    #    tobj.set_color('black')        
    fw, fh = fig.get_size_inches()
    bbox = fig.axes[2].get_position()
    aw, ah = bbox.width * fw, bbox.height * fh
    xlim = fig.axes[2].get_xlim()
    ylim = fig.axes[2].get_ylim()
    for p in fig.axes[2].patches:
        bbox1 = p.get_bbox()
        pw = bbox1.width/(xlim[1]-xlim[0]) * aw * 72
        ph = bbox1.height/(ylim[1]-ylim[0]) * ah * 72
        sw = ph/72/aw*(xlim[1]-xlim[0])
        p.set_width(sw)
    return

ana.getLance2007(5)
ana.orderData(l1, wt1)
fig = mut.plotViolinBar(ana, "CTRL vs LPS")
ana.printAllPvals()
adjustFigure(ana, fig)
pdf.savefig(transparent=True, bbox_inches = 'tight')

ana.getLance2007(13, 'CTRL')
ana.orderData(l1, wt1)
fig = mut.plotViolinBar(ana, "CTRL vs LPS")
ana.printAllPvals()
adjustFigure(ana, fig)
pdf.savefig(transparent=True, bbox_inches = 'tight')

ana.getLance2007(13, 'LPS')
ana.orderData(l1, wt1)
fig = mut.plotViolinBar(ana, "CTRL vs LPS")
ana.printAllPvals()
adjustFigure(ana, fig)
pdf.savefig(transparent=True, bbox_inches = 'tight')

bone.closePDF(pdf)
```
